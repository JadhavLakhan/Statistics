```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson,skew,kurtosis
import seaborn as sns
```

```python
# Generate a dataset (e.g., event counts in a time interval)
np.random.seed(50)  # For reproducibility
mu = 6  # Mean of the Poisson distribution
data = poisson.rvs(mu, size=1000)  # Generate random Poisson-distributed data
```

```python

# Plot the histogram of the generated data
plt.figure(figsize=(8, 5))
plt.hist(data, bins=range(0, max(data) + 2), color='lightgreen', edgecolor='black', alpha=0.7, density=True, rwidth=0.8)


# Overlay the theoretical Poisson PMF
x = np.arange(0, max(data) + 1)
pmf = poisson.pmf(x, mu)
plt.plot(x, pmf, 'o-', color='blue', label="Theoretical PMF")


# Plot settings
plt.title("Histogram of Poisson Distributed Data and Theoretical PMF", fontsize=14)
plt.xlabel("Number of Events (k)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.xticks(x)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
skewness=skew(data)
kurtos=kurtosis(data)
print(skewness)
print(kurtos)
```
